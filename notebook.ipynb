## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
dataset = pd.read_csv('sample_data/products_dataset.csv')

# Menampilkan 5 baris pertama dari dataset untuk melihat isi data
print(dataset.head())

# Menampilkan informasi tentang dataset
print(dataset.info())

# Menampilkan statistik deskriptif untuk kolom numerik
print(dataset.describe())

# Menampilkan nama-nama kolom dalam dataset
print(dataset.columns)

# Memeriksa jumlah nilai yang hilang (missing values) per kolom
print(dataset.isnull().sum())

# Memeriksa jumlah baris duplikat
duplicates = dataset.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicates}")

# Menghapus baris duplikat
dataset.drop_duplicates(inplace=True)

# Menghapus data yang hilang pada kolom 'product_weight_g' dan 'product_category_name'
dataset_cleaned = dataset.dropna(subset=['product_weight_g', 'product_category_name'])

# Analisis Pertanyaan Bisnis #5: Distribusi Berat Produk per Kategori
plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset_cleaned, x='product_category_name', y='product_weight_g')
plt.xticks(rotation=90)  # Memutar label kategori agar lebih mudah dibaca
plt.title('Distribusi Berat Produk per Kategori')
plt.xlabel('Kategori Produk')
plt.ylabel('Berat Produk (gram)')
plt.show()
```

### Pertanyaan 2:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
dataset = pd.read_csv('sample_data/products_dataset.csv')

# Menampilkan 5 baris pertama dari dataset untuk melihat isi data
print(dataset.head())

# Menampilkan informasi tentang dataset
print(dataset.info())

# Menampilkan statistik deskriptif untuk kolom numerik
print(dataset.describe())

# Menampilkan nama-nama kolom dalam dataset
print(dataset.columns)

# Memeriksa jumlah nilai yang hilang (missing values) per kolom
print(dataset.isnull().sum())

# Memeriksa jumlah baris duplikat
duplicates = dataset.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicates}")

# Menghapus baris duplikat
dataset.drop_duplicates(inplace=True)

# Menghapus data yang hilang pada kolom 'product_weight_g', 'product_length_cm', 'product_height_cm', dan 'product_width_cm'
dataset_cleaned = dataset.dropna(subset=['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm'])

# Analisis Pertanyaan Bisnis #3: Apakah produk dengan berat tinggi cenderung lebih besar?
# Scatter plot antara berat produk dan panjang produk
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset_cleaned, x='product_length_cm', y='product_weight_g')
plt.title('Hubungan antara Panjang Produk dan Berat Produk')
plt.xlabel('Panjang Produk (cm)')
plt.ylabel('Berat Produk (gram)')
plt.show()

# Scatter plot antara berat produk dan tinggi produk
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset_cleaned, x='product_height_cm', y='product_weight_g')
plt.title('Hubungan antara Tinggi Produk dan Berat Produk')
plt.xlabel('Tinggi Produk (cm)')
plt.ylabel('Berat Produk (gram)')
plt.show()

# Scatter plot antara berat produk dan lebar produk
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset_cleaned, x='product_width_cm', y='product_weight_g')
plt.title('Hubungan antara Lebar Produk dan Berat Produk')
plt.xlabel('Lebar Produk (cm)')
plt.ylabel('Berat Produk (gram)')
plt.show()
```

**Insight:**
- Beberapa kategori memiliki rata-rata berat produk yang jauh lebih besar.
- Distribusi berat produk memberikan wawasan penting untuk logistik.
